```python
#5 Month code
def find_month(x):
    if " 01 " in x:
        return "Jan"
    elif " 02 " in x:
        return "Feb"
    elif " 03 " in x:
        return "March"    
    elif " 04 " in x:
        return "April"    
    elif " 05 " in x:
        return "May"    
    elif " 06 " in x:
        return "June"    
    elif " 07 " in x:
        return "July"    
    elif " 08 " in x:
        return "August"    
    elif " 09 " in x:
        return "Sep"    
    elif " 10 " in x:
        return "Oct"    
    elif " 11 " in x:
        return "Nov"    
    else:
        return "Dec"
```

```python
df["Month"]=df["Date/Time"].apply(find_month)
```

```python
df.head()
```

```python
df.groupby("Month").mean()
```

```python
# The months with the highest average wind speed and power are March and August respectively. Yes it is exactly what I expected as the theoretical power curve is also at its highest in August, so it would make sense for actual power to be the highest in that month as well. For wind speed, I would assume that March is the windiest month due to the weather, so it makes sense for this month to have the highest mean.
```

# Part 2 - Correlation

```python
y = df["Theoretical_Power_Curve (KWh)"]
x = df["Wind Speed (m/s)"]
plt.scatter(x,y)
plt.title("Theoretical Power VS Wind Speed")
plt.ylabel("Theoretical_Power_Curve (KWh)")
plt.xlabel("Wind Speed (m/s)")
```

```python
y = df["Theoretical_Power_Curve (KWh)"]
x = df["Wind Direction (°)"]
plt.scatter(x,y)
plt.title("Theoretical Power VS Wind Direction")
plt.ylabel("Theoretical_Power_Curve (KWh)")
plt.xlabel("Wind Direction (°)")
```

```python
# 1. The Theoretical Power Curve should be on the y-axis because it is affected by how wind passes through the turbine. Depending on the wind speed and direction, determines how much power is generated. The scatter plots tell me that the power should hit a constant after a certain amount of speed and it will gradually increase in a positive linear fashion until the max.
```

```python
y = df["LV ActivePower (kW)"]
x = df["Wind Speed (m/s)"]
plt.scatter(x,y)
plt.title("Actual Power VS Wind Speed")
plt.ylabel("LV ActivePower (kW)")
plt.xlabel("Wind Speed (m/s)")
```

```python
y = df["LV ActivePower (kW)"]
x = df["Wind Direction (°)"]
plt.scatter(x,y)
plt.title("Actual Power VS Wind Speed")
plt.ylabel("LV ActivePower (kW)")
plt.xlabel("Wind Direction (°)")
```

```python
#2. The wind direction still heavily affects the power output and the plots are comparably similar to the theoretical plots.
```

```python
#3. The spread of the data is somewhat different between theoretical and actual power due to theoretical being the closest you can get to a perfect linear increase in power generation. They are different because there might be certain points in time where the wind turbine may not generate the power expected due to the wind speed and direction.
```

```python
fig = plt.figure(figsize = (9,7))
sns.heatmap(df.corr(), annot = True)
plt.show()
```

```python
#4 a) Postive correlation between Wind Speed and Theoretical Power Curve.
#  b) Negative correlation between wind Direction and Theoretical Power Curve.
#  c) Postive correlation between Wind Speed and Actual Power Curve.
#  d) Negative correlation between wind Direction and Theoretical Power Curve.

# Yes, these results agree with the four scatterplots created earlier as they show the same correlation between all 4 relationships in the graphs.
```
